```python
from ailib import multi_linear_regressor
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
net = multi_linear_regressor(size=2)
```

```python
X_train = np.array([[-2, -3],
                     [0,  0],
                     [0,  1],
                     [21, 3],
                     [1, 7],
                     [3, 4],
                     [32, 21],
                     [12, -74]])

# X_train = np.array([[-2],
#                      [0],
#                      [3],
#                      [21]])

# X_train = np.zeros((X_train.shape()))
                     
y_train = np.sum(X_train*3, axis=1)+2 # y = x1*3 + x2*3 + 2
# y_train[0] = -10
print(X_train,'\n\n', y_train)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d') # or ax = plt.axes(projection='3d')
ax.scatter3D(X_train[:,0], X_train[:,1], y_train)
ax.plot_surface
```

```python
net.predict([-2, -3])
```

```python
print(net.weights, net.bias)
net.compute_cost(X_train, y_train)
```

```python
J_history = net.gradient_descent(X_train, y_train, num_iters=100000)
```

```python
plt.plot(J_history[:100])
```
